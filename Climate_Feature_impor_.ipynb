```python
feat_rank = []
X_trans = StandardScaler().fit_transform(X)
X_trans = pd.DataFrame(X_trans, columns = X.columns)
rfe = RFE(estimator = LinearRegression(), n_features_to_select = 1)
model = LinearRegression()
pipe = Pipeline(steps=[('s' , rfe), ('m' , model)])
dfcol = pd.DataFrame(X_trans.columns)

for cols in y_cont.columns:
    pipe.fit(X_trans, np.log(y_cont[[cols]]))
    rank = pd.DataFrame(rfe.ranking_)
    feat_score = pd.concat([dfcol, rank], axis=1)
    feat_score.columns = [str(cols), 'Score']
    rankk = feat_score.nsmallest(14, 'Score')
    feat_ = rankk[str(cols)]
    feat_ = feat_.reset_index(drop=True)
    feat_rank.append(feat_)

feat_rank = pd.DataFrame(feat_rank)
feat_rank
```
